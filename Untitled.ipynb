# Dune Parser 

```python
class Dune_API:
    def __init__(self, email, password):
        self.email = email
        self.password = password

        
    def login(self):
        global dune
        dune = DuneAnalytics(self.email, self.password)
        dune.login()
        dune.fetch_auth_token()
    
    def getQueryDataByID(self,queryID):
        result_id = dune.query_result_id(query_id=queryID)
        return dune.query_result(result_id)
    
    def setUpDataFrame(self, data):
        data['data']['query_results'][0]['columns']
        global duneDF
        cols =  [data['data']['query_results'][0]['columns']][0]
        duneDF =pd.DataFrame(columns=cols)
        
    def fillDF(self, data):
        for i in range(0, len(data['data']['query_results'][0]['columns'])):
            key = duneDF.columns[i]
            for j in range(0, len( data['data']['get_result_by_result_id'])):
                duneDF.loc[j, key] = data['data']['get_result_by_result_id'][j]['data'][key]
        
                        
        
    def is_date(self,data):
        for i in range(len (duneDF.columns)):
            try:
                format_yyyymmdd = "%Y-%m-%d"
                datetime.datetime.strptime((duneDF.loc[0, duneDF.columns[i]])[:10],format_yyyymmdd)
                duneDF[duneDF.columns[i]]=  pd.to_datetime(duneDF[duneDF.columns[i]])
            except:
                pass
            
    def is_number(self):
        for i in range(len (duneDF.columns)):
            if (isinstance(duneDF[duneDF.columns[i]][0], (int, float))):
                duneDF[duneDF.columns[i]] =  pd.to_numeric(  duneDF[duneDF.columns[i]])
                
        
    
    def create_DF(self, queryID):
        self.login()
        data = self.getQueryDataByID(queryID)
        self.setUpDataFrame(data)
        self.fillDF(data)
        self.is_date(data)
        self.is_number()
        return duneDF
```

```python
def generateColors(numOfColors):
    cm = matplotlib.cm.get_cmap('Spectral')
    colors= []
    for i in range(numOfColors):
        colors.append(cm(i * (1/numOfColors)+ (1/numOfColors) ))
        #colors.append(cm(i * (1/NUM_COLORS)+ (1/NUM_COLORS) ))
    return colors
    #colors.reverse()
```

### CRV Staked in veCRV Graph 

```python
fig, ax = plt.subplots(1, figsize=(8,6))
plt.stackplot(pd.date_range(crvStakedveCRV['datevalue'].min(),crvStakedveCRV['datevalue'].max()),
    crvStakedveCRV['yearn'],
    crvStakedveCRV['stake_dao'],
    crvStakedveCRV['convex'],
    crvStakedveCRV['abracadabra'],
    crvStakedveCRV['liquity'],
    crvStakedveCRV['frax'],
    crvStakedveCRV['keep3r'],
    crvStakedveCRV['redacted'],
    crvStakedveCRV['eoas'],
    
    labels= crvStakedveCRV.columns[1:],
     
    colors = generateColors(crvStakedveCRV.shape[1] -1)
)
plt.legend(loc='upper left')
plt.ylabel('CRV', fontdict=None, labelpad=None)
plt.xlabel('Date', fontdict=None, labelpad=None)
plt.title('Total CRV Vote Locked', fontdict=None, loc='center', pad=None)
#plt.savefig('crvStakedveCRV.jpeg')
```

### Address Percentage CRV in veCRV

```python
def makePercentageCol(data):
    needed_columnms = data.columns[1:]
    df_sums = data.groupby('datevalue')[needed_columnms].sum()
    df_sums['Total'] = df_sums[needed_columnms].sum(1)
    df_sums = df_sums.reset_index(drop=False)
    
    for i in range(len(needed_columnms)):
        for j in range(data.shape[0]):
            divider = df_sums.loc[j, 'Total'] * (100/1)
            if divider != 0:
                df_sums.loc[j ,needed_columnms[i] +'percent'] =  (data.loc[j, needed_columnms[i]] / df_sums.loc[j, 'Total']) * (100/1)
    return df_sums
```

```python
crvStakedveCRVpercent = makePercentageCol(crvStakedveCRV) 
```

### Address Percentage CRV in veCRV Graph 

```python
fig, ax = plt.subplots(1, figsize=(8,6))
plt.stackplot(pd.date_range(crvStakedveCRV['datevalue'].min(),crvStakedveCRV['datevalue'].max()),
    crvStakedveCRVpercent['yearnpercent'],
    crvStakedveCRVpercent['stake_daopercent'],
    crvStakedveCRVpercent['convexpercent'],
    crvStakedveCRVpercent['abracadabrapercent'],
    crvStakedveCRVpercent['liquitypercent'],
    crvStakedveCRVpercent['fraxpercent'],
    crvStakedveCRVpercent['keep3rpercent'],
    crvStakedveCRVpercent['redactedpercent'],
    crvStakedveCRVpercent['eoaspercent'],
    
    labels= crvStakedveCRV.columns[1:],
    
    colors = generateColors(round(crvStakedveCRVpercent.shape[1]/2) -1)
)
plt.legend(loc=(1.04, .4))
plt.ylabel('%', fontdict=None, labelpad=None)
plt.xlabel('Date', fontdict=None, labelpad=None)
plt.title('Total CRV Vote Locked', fontdict=None, loc='center', pad=None)
#plt.savefig('crvStakedveCRV.jpeg')
```
